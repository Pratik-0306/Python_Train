```python
## accessing the next blank cell in sheet and updating the next blank row with new input
def next_blank_cell(worksheet):
  string=list(filter(None,worksheet.col_values(1)))
  return str(len(string)+1)
blank=next_blank_cell(worksheet)
print(blank)
```
